### While Loops

```python
hearts = 10
mode = "survival"
print("mode =", mode)
while mode=="survival":
    print("Hearts =", str(hearts))
    hearts -= 1
    if hearts < 1:
        mode = "creative"
print("mode =", mode)
```

##### Looping 'While True'

```python
game_over = True

while game_over:
    print("Infinite Loop")
    user_input = input("Would you like to quit? ")
    if user_input == "yes":
        game_over = False

```

##### While & For Loops Used Together

```python
num = 0

while num < 5:
    print("\n While loop ineration: " + str(num))
    
    for i in range(2):
        print("For loop interation:", str(i))
    
    num += 1
```

##### Looping Through Lists

```python
blocks = ["grass_block", "dirt", "gravel", "sand", "stone", "planks", "cobblestone"]

print("There are " + str(len(blocks)) + " blocks")

for block in range(len(blocks)):
    print(blocks[block])

for block in blocks:
    print(block)
```

## Exercise #2 <br>
<p>Get first prime numbers up to 100</p>

```python
# HINT::
# An else after an if runs if the if didnt
# An else after a for runs if the for didnt break

print("Here are a list of prime numbers smaller than 100")
for i in range(1, 101, 1):
    k=0
    if (i%1 == 0 & i%i == 0):
        for j in range(2, i, 1):
            if (i%j == 0):
                k+=1
    if (k < 1):
        print(i)
        
            
    
    
```
